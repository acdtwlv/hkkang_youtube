## 파일 로딩

파일을 열어서 조사해 보겠습니다. Pandas의 read_excel() 이란 함수를 이용하면 열 수 있네요.

```python
import pandas as pd
```

```python
filepath = 'QuantKing.xlsx'
df = pd.read_excel(filepath, sheet_name='퀀트데이타', skiprows=2).drop('Unnamed: 0', axis=1)
df.head()
```

강의에서 사용하는 컬럼만 잘라서 보관하도록 하겠습니다.

```python
df= df[['회사명', '시가총액\n(억)', 
        '1개월\n등락률\n(%)', '3개월\n등락율\n(%)', '6개월\n등락율\n(%)', '9개월\n등락율\n(%)', '1년\n등락율\n(%)', # 모멘텀
        '발표\nPBR', '발표\nPER', '1년후\nPER', '과거\nEV/EBITDA\n(%)', '과거\nPSR',# 가치
        '과거\nROA\n(%)', '단순\n부채비율\n(%)', '14년->17년\n3년간 YOY', '배당\n성향\n(%)', '주가\n변동성' # 우량주
       ]].set_index('회사명')
df.head()
```

## 모멘텀 순위

<span style="color:red">**한국에서 모멘텀은 대형주에서만 통하기에, 시가총액 기준 상위 200개 기업만 분석합니다.**</span> 

```python
df = df.sort_values(by='시가총액\n(억)', ascending=False).iloc[:200]
df.head()
```

최근 한달의 모멘텀은 없애주고, 등수륵 구해서 평균해줍니다.

```python
df['3-1 모멘텀'] = df['3개월\n등락율\n(%)'] - df['1개월\n등락률\n(%)']
df['6-1 모멘텀'] = df['6개월\n등락율\n(%)'] - df['1개월\n등락률\n(%)']
df['9-1 모멘텀'] = df['9개월\n등락율\n(%)'] - df['1개월\n등락률\n(%)']
df['12-1 모멘텀'] = df['1년\n등락율\n(%)'] - df['1개월\n등락률\n(%)']

df['3순위'] = df['3-1 모멘텀'].rank(ascending=False)
df['6순위'] = df['6-1 모멘텀'].rank(ascending=False)
df['9순위'] = df['9-1 모멘텀'].rank(ascending=False)
df['12순위'] = df['12-1 모멘텀'].rank(ascending=False)

df['모멘텀 통합순위'] = df[['3순위', '6순위', '9순위', '12순위']].mean(axis=1)

df.sort_values(by='모멘텀 통합순위').head()
```

강의에서 본 것처럼 아난티가 모멘텀이 가장 좋은 기업입니다.

## 가치 순위

가치 순위를 매겨보겠습니다.

<span style="color:red">**간혹 PER, PBR 등이 없는 주식이 나옵니다. 그냥 지워버리면 됩니다.**</span>

```python
df = df[df['발표\nPBR'] != 0]
df = df[df['발표\nPER'] != 0]
df = df[df['1년후\nPER'] != 0]
df = df[df['과거\nEV/EBITDA\n(%)'] != 0]
df = df[df['과거\nPSR'] != 0]

df['1/PBR'] = 1 / df['발표\nPBR']
df['1/PER'] = 1 / df['발표\nPER']
df['1/1년후 PER'] = 1 / df['1년후\nPER']
df['1/EVEBITDA'] = 1 / df['과거\nEV/EBITDA\n(%)']
df['1/PSR'] = 1 / df['과거\nPSR']

df['순위 PBR'] = df['1/PBR'].rank(ascending=False)
df['순위 PER'] = df['1/PER'].rank(ascending=False)
df['순위 1년후 PER'] = df['1/1년후 PER'].rank(ascending=False)
df['순위 EVEBITDA'] = df['1/EVEBITDA'].rank(ascending=False)
df['순위 PSR'] = df['1/PSR'].rank(ascending=False)
df['통합 가치 지표'] = df[['순위 PBR', '순위 PER', '순위 1년후 PER', '순위 EVEBITDA', '순위 PSR']].mean(axis=1)

df.sort_values(by='통합 가치 지표').head()
```

## 우량주 순위

우량주 순위를 매겨보겠습니다.

<span style="color:red">**중요: 부채비율/변동성은 낮은게 좋습니다.**</span>

```python
df['순위 ROA'] = df['과거\nROA\n(%)'].rank()
df['부채비율 순위'] = df['단순\n부채비율\n(%)'].rank(ascending=True)
df['성장률 순위'] = df['14년->17년\n3년간 YOY'].rank(ascending=False)
df['배당성향 순위'] = df['배당\n성향\n(%)'].rank(ascending=False)
df['변동성 순위'] = df['주가\n변동성'].rank(ascending=True)

df['우량주 통합순위'] = df[['순위 ROA', '부채비율 순위', '성장률 순위', '배당성향 순위', '변동성 순위']].mean(axis=1)

df.sort_values(by='우량주 통합순위').head()
```

## 통합 순위

통합 순위를 매겨보겠습니다. 

<span style="color:red">**비중: 저평가 지표 * 0.4 + 모멘텀 지표 * 0.4 + 우량주 지표 * 0.2**</span>

```python
df['통합순위'] = df['모멘텀 통합순위'] * 0.4 + df['통합 가치 지표'] * 0.4 + df['우량주 통합순위'] * 0.2
```

```python
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
df.sort_values(by='통합순위').head()
```

강의에서처럼 대림산업이 1위를 차지했습니다. 강의에서 2위한 KT가 여기서는 3위했구요.

잡담1: 동영상을 보면 에러가나서 회사 하나하나 살펴보는 부분이 있습니다. 퀀트킹 데이터뿐 아니라 세상의 모든 데이터에는 오류가 있을 수 있습니다. 결과가 이상해보이면 언제든 결과분석의 어느 부분이던 (처음 자료부터) 의심하는 습관을 가져야 함을 잘 알려주셨습니다.
